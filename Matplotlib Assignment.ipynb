A scatter plot is a type of graph that shows the relationship between two variables. Each data point in a scatter plot is represented by a dot, and the position of the dot on the graph indicates the values of the two variables for that data point.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.title("Scatter Plot")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.scatter(x,y)
plt.show()
```

```python
#3
#The subplot() function is used to create multiple plots in a single figure. It is a convenient way to visualize multiple datasets or to compare different results.
```

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550,450,150])

plt.subplot(2, 3, 4)
plt.plot(x,y)

plt.show()
```

A barplot is one of the most common types of graphic. It shows the relationship between a numeric and a categoric variable.
A bar diagram makes it easy to compare sets of data between different groups at a glance. The graph represents categories on one axis and a discrete value in the other.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```

A box plot, also known as a box-and-whisker plot, is a graphical way of displaying the distribution of numerical data. It shows five numbers: the minimum, first quartile (Q1), median, third quartile (Q3), and maximum. The box plot is a standardized way of displaying the dataset based on the five-number summary.

It is used for :-

Visualize the distribution of data. 

Compare different sets of data.

Identify outliers.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
plt.show()
```

```python
plt.boxplot(box2)
plt.show()
```
